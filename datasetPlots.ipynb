```python
import os
import matplotlib.pyplot as plt
%matplotlib inline 
# everyone will thank me for this
plt.style.use('dark_background')
```

```python
classNumbers = {"0" : 0, "1":0}
```

```python
datasetDir = "AugmentedTreesDatasetTwoClass/"
dsSplit = ["test", "train", "valid"]

for split in dsSplit:
    subPath = os.path.join(datasetDir, split)
    
    for labelFile in os.listdir(os.path.join(subPath, "labels")):
        labelPath = os.path.join(subPath, "labels", labelFile)
        
        lines = []
        # print(labelPath)
        with open(labelPath) as f:
            lines = f.readlines()
        
        for line in lines:
            classNumbers[line[0]] += 1
                    
```

```python
classes = ["banana", "coconut"]
values = list(classNumbers.values())

 
fig = plt.figure(figsize = (5, 5))

plt.bar(classes, values, color ='aqua', 
        width = 0.2)

plt.xlabel("Class Names")
plt.ylabel("Instances")
plt.title("Class Distribution")
```
