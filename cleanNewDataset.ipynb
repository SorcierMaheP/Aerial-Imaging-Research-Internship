Notebook to remove all the smaller calsses in the datset, keeping only coconut and banana


```python
import os
```

```python
datasetDir = "AugmentedTreesDatasetTwoClass/"
dsSplit = ["test", "train", "valid"]

for split in dsSplit:
    subPath = os.path.join(datasetDir, split)
    
    for labelFile in os.listdir(os.path.join(subPath, "labels")):
        labelPath = os.path.join(subPath, "labels", labelFile)
        
        lines = []
        # print(labelPath)
        with open(labelPath) as f:
            lines = f.readlines()
        
        for line in lines:
            if line[0] not in ['0','1']:
                # print(line)
                os.remove(labelPath)
                # print(labelPath)
                imageName = labelFile[:-4] + ".jpg"
                imagePath = os.path.join(subPath, "images", imageName )
                os.remove(imagePath)
                break
                # print(imagePath)
                    
```
